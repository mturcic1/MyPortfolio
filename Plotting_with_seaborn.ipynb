```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (12,8)
```

```python
mass_shootings = pd.read_csv("mass_shootings.csv")
```

```python
mass_shootings.head()
```

```python
mass_shootings['Incident Date'] = pd.to_datetime(mass_shootings['Incident Date'])
```

```python
mass_shootings.dtypes
```

```python
state_counts = mass_shootings.groupby('State')['Incident ID'].count().reset_index()
state_counts = state_counts.sort_values(by='Incident ID', ascending=False)
avg_incidents = state_counts['Incident ID'].mean()
# plotting
ax = sns.barplot(x='State', y='Incident ID', data=state_counts, color="green")
plt.title('Number of mass shootings per state')
plt.xlabel('State')
plt.ylabel('Number of Incidents')
# rotating x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# adding mean line
plt.axhline(y=avg_incidents, color='r', linestyle='--')
```

```python
city_counts = mass_shootings.groupby('City Or County')['Incident ID'].count().reset_index().head(20)
city_counts = city_counts.sort_values(by='Incident ID', ascending=False)
avg_incidents = city_counts['Incident ID'].mean()
# plotting
ax = sns.barplot(x='City Or County', y='Incident ID', data=city_counts, color="blue")
plt.title('Cities and districts with highest number of mass shootings')
plt.xlabel('City or district')
plt.ylabel('Number of Incidents')
# rotating x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# adding mean line
plt.axhline(y=avg_incidents, color='r', linestyle='--')
```

```python
state_counts = mass_shootings.loc[mass_shootings['# Killed'] > 0].groupby('State')['Incident ID'].count().reset_index()
state_counts = state_counts.sort_values(by='Incident ID', ascending=False)
avg_incidents = state_counts['Incident ID'].mean()
# plotting
ax = sns.barplot(x='State', y='Incident ID', data=state_counts, color="green")
plt.title('Number of Fatal mass shootings by State')
plt.xlabel("State")
plt.ylabel('Number of Incidents')
# rotating x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# adding mean line
plt.axhline(y=avg_incidents, color='r', linestyle='--')
```

```python
city_counts = mass_shootings.loc[mass_shootings['# Killed'] > 0].groupby('City Or County')['Incident ID'].count().reset_index().head(20)
city_counts = city_counts.sort_values(by='Incident ID', ascending=False)
avg_incidents = city_counts['Incident ID'].mean()
# plotting
ax = sns.barplot(x='City Or County', y='Incident ID', data=city_counts, color="blue")
plt.title('Cities with highest number of fatal mass shootings')
plt.xlabel('City or district')
plt.ylabel('Number of Incidents')
# rotating x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# adding mean line
plt.axhline(y=avg_incidents, color='r', linestyle='--')
```

```python
state_counts = mass_shootings.groupby('State')['# Killed'].sum().reset_index()
state_counts = state_counts.sort_values(by='# Killed', ascending=False)
avg_incidents = state_counts['# Killed'].mean()
# plotting
ax = sns.barplot(x='State', y='# Killed', data=state_counts, color="green")
plt.title('Number of fatal casualties by State')
plt.xlabel('State')
plt.ylabel('Number of fatal casualties')
# rotating x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# adding mean line
plt.axhline(y=avg_incidents, color='r', linestyle='--')
```

```python
city_counts = mass_shootings.groupby('City Or County')['# Killed'].sum().reset_index()
city_counts = city_counts.sort_values(by='# Killed', ascending=False).head(20)
# plotting
ax = sns.barplot(x='City Or County', y='# Killed', data=city_counts, color="blue")
plt.title('Cities and districts with highest number of fatal casualties')
plt.xlabel('City or district')
plt.ylabel('Number of fatal casualties')
# rotating x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# adding mean line
plt.axhline(y=avg_incidents, color='r', linestyle='--')
```

```python
state_counts = mass_shootings.groupby('Incident Date')['Incident ID'].count().reset_index()

# plotting
ax = plt.subplots(figsize=(20, 5))
ax = sns.lineplot(x='Incident Date', y='Incident ID', data=state_counts, color="green")
plt.title('Number of mass shootings troughout the years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
```
